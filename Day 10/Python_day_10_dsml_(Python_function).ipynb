3. No parameter , has return type
4. Has parameter , has return type

```python
# 1. No parameter , no return type

def greet():
  print("hello! welcome to skillshikshya")
```

```python
#calling greet function

greet()
```

```python
# 2. No parameter , has return type

def get_pi():
  pi = 3.14
  return pi
```

```python
value = get_pi()
print("the value of pi is ,value")
```

```python
# 3. Has parameter , no return type

def greet(name):
  print(f"hello, {name} !")  # f in front of string is used to put value in between trhe string
```

```python
greet("Nelson")
```

```python
# 4. Has parameter , has return type

def add(a,b):
  sum = a+b
  return sum

# function call
add(5,6)
```

```python
# Class Task 2

def data():
    a = int(input("Enter the first number: "))
    b = int(input("Enter the second number: "))
    op = input("Enter the operator (+, -, *, /): ")

    if op == '+':
        return a + b

    elif op == '-':
        return a - b

    elif op == '*':
        return a * b

    elif op == '/':
        if b != 0:
            return a / b
        else:
            print("Error: Division by zero!")
            return None

    else:
        print("Invalid operator. Please use +, -, *, or /.")
        return None


result = data()

if result is not None:
    print("Result =", result)
```

```python
data()
```

```python
#class task 3
def lst_operation():
  number = []

  for i in range(5):
    value = input(f"Enter value {i+1} = ")
    if "." in value :
      if len(number) > 0:
        removed = number.pop()
        print(f"float detected. removed: {removed}")
      else:
        print("float detected, but the list is empty")
    else:
      number.append(int(value))
      print(f"Integer {value} added to the list")
      print("updated List :",number)
```

```python
lst_operation()
```
