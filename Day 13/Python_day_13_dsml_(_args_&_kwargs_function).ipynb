#*args function

```python
def sum(sum):
  result = 0
  result += sum
  print(result)
```

```python
sum(1)
```

```python
def sum(*num):
  result = 0
  for n in num:
    result += n
  print(result)
```

```python
sum(1,2,4,5)
```

```python
def add(*args):
  return sum(args)
```

```python
sum(1,2,3)
```

#**kwargs function

```python
#def function(**kwargs):

#s1 = 'python'  # used for sending all value
#s2 = 'is'
#s3 = 'nice'


def function(**kwargs):
  for key,value in kwargs.items():
    print(key, "=" , value)
```

```python
function(s1 = 'python', s2 = 'is', s3 = 'nice')
```

```python
def introduce(**kwargs):
  details = []
  for k , v in kwargs.items():
    print(k,":",v)
    details.append(k + ":" + str(v))
  return ",".join(details)
```

```python
introduce(name = "alice",age = "25",city = "ktm")
```

#class task - 1

Write a function total_salary(*args) that accepts an employee's salary components(basic salary, allowance, bonus, etc.) and returns the total salary.

```python
def total_salary(*args):
  total = 0
  for salary in args:
    total += salary
    return total
```

```python
total_salary(100000,20000,30000) #(basic_salary,bonus,allowance)
```

#class task-2

Create a function cart_total(*args) that accepts the prices of any number of
products and returns the total bill amount.

```python
def cart_total(*args):
    total = 0
    for products in args:
      total += products
    return total

```

```python
cart_total(1000,20000,30000)
```

#class task-4

Create a function product_info(**kwargs) that accepts product details such as name,price, brand, and category, then displays them in a readable format.

```python
def product_info(**kwargs):
  product = []
  for k , v in kwargs.items():
    product.append(k + ":" + str(v))
  return ",".join(product)
```

```python
product_info(name="fridge",price=20000,brand="LG",category="machine")
```
