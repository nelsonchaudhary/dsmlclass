#diff between pop in list or tuple and dictionary

```python
#difference between pop in list or tuple and dictionary

lst = [1,2,3,4,5]
lst.pop()
```

```python
#in list or tuple pop removes the last value
lst
```

```python
dict_={
    "name" : "nelson2",
    "address" : "kathmandu",
    "phone" : 989795864,

}
```

```python
#in dictionary it does not remove last value by default it needs an value
dict_.pop()
```

```python
dict_.pop("name")
```

```python
dict_
```

```python
# in dictionary we can leve comment is the value to be deleted is not found
dict_.pop("name","not found")
```

#**kwargs

```python
dict1 = dict({
    "name" : "nelson2",
    "address" : "kathmandu",
})
dict1
```

```python
dict2 = dict({
    "phone" : 989795864,
    "age" : 22
})
dict2
```

```python
dict3 = dict({
    "gender":"male",
    "marks":90
})
dict3
```

```python
student_data = {**dict1, **dict2, **dict3}
student_data
```

```python
result= sorted(student_data)
result
```

#merge

```python
#merge
dict1 | dict2 | dict3
```

```python
#length(gives the number of keys)
len(student_data)
```

#setdefault()

```python
dict_data = dict({
    "name" : "nelson",
    "address" : "kathmandu",
    "phone" : 989795864,
    "course" : ["python","java","c++"]
})
dict_data
```

```python
#returns the value which is in dictionary
dict_data.setdefault("name","hello")
```

```python
# add key and value if the key and value does not exits
dict_data.setdefault("age",20)
```
