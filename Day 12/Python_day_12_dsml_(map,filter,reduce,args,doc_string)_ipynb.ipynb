#enumerate() and zip()

```python
data1 = ["ram","shyam","gita","rita"]
data2 = [20,30,40,50]

for i in zip(data1,data2):
  print(i)
```

```python
for i,(name,age) in enumerate(zip(data1,data2)):
  print(i,name,age)
```

#find maximum number using reduce() function

```python
from functools import reduce
numbers = [12, 45, 7, 89, 23, 56]

def find_max(x,y):
  if x>y:
    return x
  else:
    return y

maximum_number = reduce(find_max, numbers)
maximum_number
```

#class task
#(2D) area = l * w
#(3D) area = l * b * h
#solve using the position argument

```python
def area(*arg):
    if len(arg) == 2:
        l, w = arg
        print("2D area =" , l * w)
    elif len(arg) == 3:
        l, b, h = arg
        print("3D volume =" , l * b * h)
    else:
      print("invalid number of argument")
```

```python
area(10, 5)
```

```python
area(10, 5, 2)
```

```python
area(2)
```

```python
# Multiply all numbers using *args
def mux(*args):
    result = 1
    for num in args:
        result *= num
    return result

```

```python
print(mux(2, 3, 4))      # 24
print(mux(5, 10))         # 50
print(mux(1, 2, 3, 4, 5)) # 120
```
