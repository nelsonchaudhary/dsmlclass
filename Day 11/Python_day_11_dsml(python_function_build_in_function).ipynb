```python
#class task 1

def number(num1,num2,num3):

  if num1 > num2 and num1 > num3 :
    print("the largest number is ",num1)
  elif num2 > num1 and num2 > num3 :
    print("the largest number is ",num2)
  else:
    print("the largest number is ",num3)
```

```python
number(10,20,30)
```

```python
#class task 2
#counting vowels in text

def count_vowels(text):
  vowels = "aeiouAEIOU"
  count = 0

  for char in text:
    if char in vowels:
      count +=1
  return count
```

```python
count_vowels("umbrella")
```

ENUMERATE

```python
#enumerate
for i in range(6):
  print(i)
```

```python
lst = [1,2,3,4,5,6,7]
#enumerate
for index,num in enumerate(lst):
  print(num)
```

```python
names = ["ram","hari","sita"]
for (i,name) in enumerate(names):
  print(f"{name} : {i}")
```

```python
names = ("ram","hari","sita")
for (i,name) in enumerate(names,start =1):
  print(f"{name} : {i}")
```

ZIP

```python
#zip
data1 = [1,2,3,4,5]
data2 = ["A","B","C","D","E"]

for (data,num) in zip(data1,data2):
  print(data,num,sep="->")
```
