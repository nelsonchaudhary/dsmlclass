#SHALLOW COPY VS DEEP COPY<br>
 Shallow Copy<br>
"Copy the bag, keep the same items."<br>
 If you change an item, both bags show the change.<br>
Shallow = Share<br>

Deep Copy<br>
"Copy the bag and copy every item inside."<br>
 If you change an item, only that bag changes.<br>
Deep = Duplicate<br>

![Screenshot 2026-07-10 1206584.png](<attachment:Screenshot 2026-07-10 1206584.png>)

```python
lst = [[1,2,3,4],[5,6,7]]
lst1 = lst
print(lst)
print(lst1)
```

```python
lst2 = lst1.copy()
lst2
```

```python
lst3 = [[1,2,3,4],[5,6,7]]
lst4 = lst3.copy()
```

```python
lst3[1][1] = 100
print("original")
print(lst3)
print("copy")
print(lst4)
```

```python
import copy
lst5 = [[1,2,3,4],[5,6,7]]
lst6 = copy.deepcopy(lst5)
lst5[1][2] = 100                               #[(0,0) (0,1) (0,2)]
print("original")                              #[(1,0) (1,1) (1,2)]
print(lst5)
print("copy")
print(lst6)
```

#HOW STACK WORK<br>
a list as a stack, where the last
element added is the first element retrieved (“last-in, first-out”). 
To add an item to the top of the stack, use append(). To retrieve an item 
from the top of the stack, use pop() without an explicit index.
