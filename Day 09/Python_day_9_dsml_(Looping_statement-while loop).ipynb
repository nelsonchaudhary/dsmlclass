#break

```python
count = 1
while count < 10:

  if count == 3 :
    break

  print(count)
  count += 1
print("loop ended")
```

#continue

```python
count = 1
while count < 10:
  if count == 3:
    count += 1
    continue
  print(count)
  count += 1
print("loop ended")
```

#pass

```python
count = 1
while count < 10:
  if count == 3:
    count += 1
    pass
  print(count)
  count += 1
print("loop ended")

```

#atm machine class task

```python
# ATM Machine

balance = 20000
pin = 2587

while True:
    pin1 = int(input("Enter PIN: "))

    if pin1 == pin:
        while True:
            print("\nChoose Services")
            print("1. Check Balance")
            print("2. Deposit")
            print("3. Exit")

            choice = int(input("Enter your choice: "))

            if choice == 1:
                print("Your balance is:", balance)

            elif choice == 2:
                amount = int(input("Enter amount to be deposited: "))
                balance += amount
                print("Deposit successful.")
                print("Your new balance is:", balance)

            elif choice == 3:
                print("Thank you for using our ATM!")
                break

            else:
                print("Invalid choice. Please try again.")

        break

    else:
        print("Incorrect PIN. Try again.")
```
