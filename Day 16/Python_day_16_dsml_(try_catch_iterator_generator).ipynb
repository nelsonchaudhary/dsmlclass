```python
def count_up(n):
  i = 1
  while i<=n:
    yield i    # yield - pause here and return a value,but remember the current state of the function
    i +=1
```

```python
gent = count_up(3)
```

```python
gent
```

```python
print(next(gent))
```

```python
#try:
#expect
#else:
#finally:

try:
  numerator = int(input("Enter a number to divide : "))
  denominator = int(input("Enter a number to divide by :"))
  result = numerator/denominator
except ZeroDivisionError:
  print("Error : Cannot divide by zero!")
except ValueError:
  print("Error : Invalid input . Please enter a valid number")
else:
  print(f"the result is {result}")
finally:
  print("Execution complete.")
```

#Error raise

```python
def set(age):
  if age < 0:
    raise ValueError("age cannot be negative")
  print(f"Age set to {age}")
```

```python
try:
  set(-5)
except ValueError as e:
  print(e)
```

#custom Expection<br>  
self made error type




```python
#custom Expection

class InvalidAgeError(Exception):
  """Custom expection for invalid age values"""
  pass

def set_age_custom(age):
  if age>120:
    raise InvalidAgeError("age cannot be over 120")
```

```python
set_age_custom(140)
```
