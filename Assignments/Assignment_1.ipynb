```python
#Create a Python program using at least 10 string methods from the table above
#and show their outputs with examples. Try methods like upper(), split(),
#replace(), find(), and strip().

#Take a sentence from the user and perform different string operations on it
#such as converting case, counting words, checking digits/alphabet, and
#replacing words. Print all results clearly with labels.

# Taking input from the user
sentence = input("Enter a sentence: ")


# 1. upper()
print("1. Upper Case :", sentence.upper())

# 2. lower()
print("2. Lower Case :", sentence.lower())

# 3. title()
print("3. Title Case :", sentence.title())

# 4. capitalize()
print("4. Capitalized :", sentence.capitalize())

# 5. strip()
print("5. Strip Spaces :", sentence.strip())

# 6. split()
words = sentence.split()
print("6. Split into Words :", words)

# 7. Count words
print("7. Word Count :", len(words))

# 8. replace()
old_word = input("\nEnter a word to replace: ")
new_word = input("Enter the new word: ")
print("8. Replaced Sentence :", sentence.replace(old_word, new_word))

# 9. find()
search_word = input("\nEnter a word to find: ")
position = sentence.find(search_word)

if position != -1:
    print("9. Word Found At Index :", position)
else:
    print("9. Word Not Found")

# 10. startswith()
start = input("\nEnter a word to check if the sentence starts with it: ")
print("10. Starts With :", sentence.startswith(start))

# 11. endswith()
end = input("Enter a word to check if the sentence ends with it: ")
print("11. Ends With :", sentence.endswith(end))

# 12. isalpha()
print("\n12. Is Alphabet Only? :", sentence.isalpha())

# 13. isdigit()
print("13. Is Digit Only? :", sentence.isdigit())

# 14. isalnum()
print("14. Is Alphanumeric? :", sentence.isalnum())

# 15. count()
count_word = input("\nEnter a word to count: ")
print("15. Occurrences :", sentence.count(count_word))
```
