#upper()

```python
data = " i bought 3 apples and 2 orange for my family yesterday happily"
data.upper()
```

#lower()

```python
data = " I BOUGHT 3 APPLES AND 2 ORANGES FOR MY FAMILY YESTERDAY HAPPILY"
data.lower()
```

#title()

```python
data = " i bought 3 apples and 2 orange for my family yesterday happily"
data.title()
```

#capitalize()

```python
data = " i bought 3 apples and 2 orange for my family yesterday happily"
data.capitalize()
```

#strip()

```python
data = "       i bought 3 apples and 2 orange for my family yesterday happily        "
data.strip()
```

#split()

```python
data = "bike,scooter,car"
data.split("e")
```

```python
data =input("enter a sentence = ")
word = input("enter a word to be replaced = ")

word1 = input("enter a word to replace with = ")
data.replace(word,word1)
```
