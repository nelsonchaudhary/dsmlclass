#Student Result Management System

Create a user-defined function called student_result() that:<br>
1.Takes the names and marks of 5 students as input.<br>
2.Stores student data in a dictionary.<br>
3.Uses conditional statements to assign grades:<br>
A: 80 and above<br>
B: 60–79<br>
C: 40–59<br>
F: Below 40<br>
4.Displays each student's name, marks, and grade.<br>
5.Calculates and displays:<br>
Highest marks<br>
Lowest marks<br>
Average marks

```python
def student_result():
  student = {}

  for i in range(5):
    name = input(f"Enter the name of student {i+1} = ")
    marks = float(input(f"Enter the marks obtained by {name} = "))
    student[name] = marks
  print("Student result")
  for name , marks in student.items():
    if marks >= 80:
      grade = "A"
    elif marks >= 60:
      grade = "B"
    elif marks >= 40:
      grade = "C"
    else :
      grade = "F"
    print(f"Name = {name}")
    print(f"Marks = {marks}")
    print(f"Grade = {grade}")
    print()
    mark_list = list(student.values())
    highest = max(mark_list)
    lowest = min(mark_list)
    average = sum(mark_list) / len(mark_list)
    print("highest marks is ",highest)
    print("lowest marks is ",lowest)
    print("average marks is ",average)

#function call
student_result()
```

#Smart Shopping Cart System
Create a user-defined function called shopping_cart() that:<br>
1.Allows the user to enter 5 product prices.<br>
2.Stores the prices in a list.<br>
3.Uses a loop to calculate the total bill.<br>
4.Applies discounts using conditional statements:<br>
Total ≥ 5000 → 15% discount<br>
Total ≥ 3000 → 10% discount<br>
Total ≥ 1000 → 5% discount<br>
Otherwise → No discount<br>
5.Displays:<br>
Product list<br>
Total amount<br>
Discount amount<br>
Final payable amount



```python
def shopping_cart():
    products = []

    for i in range(5):
        price = float(input(f"Enter the price of product {i+1}: "))
        products.append(price)

    # total bill
    total = 0
    for price in products:
        total += price

    if total >= 5000:
        discount = total * 0.15
    elif total >= 3000:
        discount = total * 0.10
    elif total >= 1000:
        discount = total * 0.05
    else:
        discount = 0

    final_amount = total - discount


    print("Product Prices:", products)
    print("Total Amount =", total)
    print("Discount Amount =", discount)
    print("Final Payable Amount =", final_amount)


# Function Call
shopping_cart()
```
