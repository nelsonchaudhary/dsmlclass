#PYTHON SET OPERATION
#union (|)
#intersection (&)
#difference (-)
#symmetric difference (^)

```python
#union
data = {1,2,3,4,5}
data1 = {3,4,5,6,7}
data | data1
```

```python
#intersection
data = {1,2,3,4,5}
data1 = {3,4,5,6,7}
data & data1
```

```python
#difference
data = {1,2,3,4,5}
data1 = {3,4,5,6,7}
print(data - data1)
print(data1 - data)
```

```python
#symmetric difference
print((data-data1) | (data1-data))
```

#OR

```python
#symmetric difference
data = {1,2,3,4,5}
data1 = {3,4,5,6,7}
data ^ data1
```
