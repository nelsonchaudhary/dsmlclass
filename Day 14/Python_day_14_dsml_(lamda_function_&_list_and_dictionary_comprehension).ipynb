```python
#class work
#using reduce() : sum of all numbers -> using lambda

from functools import reduce

lst = [1,2,3,4,5,6,7,8,9]

sum = reduce(lambda x,y:x+y,lst)
sum
```

#python list comprehension

```python
data1 = ["welcome","to","dsml","class"]
for word in data1:
  if 'e' in word:
    print(word.upper())
```

```python
[word.upper() for word in data1 if 'e' in word]
```

```python
#numbers = [1,2,3,4,5,6]
#output
#['odd','even,...']


numbers = [1,2,3,4,5,6]

[ "even" if num%2==0 else "odd" for num in numbers]
```

```python
number = [1,2,3,4,5,6]

even = [num for num in number if num%2==0]
print(even)
```

#list comprehension

```python
list1 = [1, 2]
list2 = ['A', 'B']

pairs = [(x, y) for x in list1 for y in list2]

print(pairs)
```

#dictionary comprehension with condition<br>
{key:value for var in iterable}

```python
for num in range(1,11):
  print({num :num*num})
```

```python
numbers = [1,2,3,4,5,6]
{num: num**2 for num in numbers if num%2==0}
```

```python
{num: num**2 for num in range(1,11)}
```

```python

original_dict = {'ram':38,'sita':45,'gita':57,'jhon':34}
even_dict = {k:v for (k,v) in original_dict.items() if v%2==0}     #k:v is key:value
even_dict
```

```python
#add the condition
#1. value is not even
#2.value greater than 40


original_dict = {'ram':38,'sita':45,'gita':57,'jhon':34}
even_dict = {k:v  for (k,v) in original_dict.items() if v%2!=0 if v>40 }
even_dict
```

```python
#numbers = [1,2,3,4,5,6]
#output
#['odd','even,...']


numbers = [1,2,3,4,5,6]

result = {
    num : "even" if num%2==0 else "odd" for num in numbers
}
print(result)
```

#class task-5
Given a list of student names, use list comprehension to select names with more than 4 characters and dictionary comprehension to store each selected name with its character count.

```python
students = ["John", "Alice", "Bob", "Michael", "Emma", "Sophia", "Tom"]
```

```python
selected_names = [name for name in students if len(name) > 4]
selected_names
```

```python
name_length = {name: len(name) for name in selected_names}
name_length
```
