#AND OR OPERATORS

```python
#and operator
#true + ture +true otherwise false

age = 20
citizenship = True

if age >=18 and citizenship:
  print("eligible to vote")
else:
  print("not eligible")
```

```python
#OR operator
#false + false = false otherwise true

username = "admin"
password = "1234"

if username == "admin" or password == "12345":
  print("logged in")
else:
  print("not logged in")
```

#any() function with if

```python
marks =[25,30,40,10]
if any(mark<35 for mark in marks):
  print("at least one student passed")
```

```python

admin =False
teacher =True

if any([admin , teacher]):
  print("Access granted")
```

all() function with if

```python
marks =[60,30,40,50]
if any(mark>=30 for mark in marks):
  print("All student passed")
```
