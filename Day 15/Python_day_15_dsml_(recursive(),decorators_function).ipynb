```python
# factorial number

# 5! = 5*4*3*2*1 = 120
# 7! = 7*6*5!

# 1) n* (n-1) (recursive case)
# 2) 1 or 0 (base case)
```

```python
def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n*factorial(n-1)
```

```python
factorial(5)
```

```python
#recursive fibonacci series.

def recur_fibo(n):
    if n <= 1:
        return n
    else:
        return recur_fibo(n - 1) + recur_fibo(n - 2)

n = int(input("Enter a number: "))

if n <= 0:
    print("Please enter a positive integer.")
else:
    print("Fibonacci sequence:")
    for i in range(n):
        print(recur_fibo(i))
```

#decorators

```python
#nested function

def outer(x):
  def inner(y):
    return x+y
  return inner
```

```python
add_five = outer(5)
```

```python
add_five
```

```python
result = add_five(6)
result
```

```python
#pass function as argument
def add(x,y):
  return x+y
def calculate(func,x,y):
  return func(x,y)
```

```python
result = calculate(add,4,5)
result
```

```python
# return a function as a value.

def greeting(name):
  def hello():
    return "hello," + name + "!"
  return hello
```

```python
greet = greeting("Nelson")
```

```python
greet
```

```python
greet()
```

```python
def myWrapper(func):
  def myinnerfunc():
    print("inner wrapper.")
    func()
  return myinnerfunc
```

```python
@myWrapper
def myFunc():
  print("Hello,world!")
```

```python
myFunc()
```

```python
def logger(func):
  def wrapper(*args,**kwargs):
    print(f"calling {func.__name__}")
    return func(*args,**kwargs)
  return wrapper
```

```python
@logger
def greet(name,age):
  print(f"hello {name}, Age: {age}")
```

```python
greet("nelson",23)
```
